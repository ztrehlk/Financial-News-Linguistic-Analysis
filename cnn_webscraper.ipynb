### Article Parser Function
- note: this function should work for any page and will return anything designated as a paragraph (or <p>) in lowercase list format
- the list is a list of each word included

```python
def article_parser(url):
    import requests
    from bs4 import BeautifulSoup

    #http request
    r = requests.get(url)

    #get html format
    soup = BeautifulSoup(r.text, "html.parser")

    #finds all the 
    html_list = list(soup.find_all("p"))
    lowercase_letter_list = []


    for item in html_list:
        temp = item.text.strip().split(' ')
        for word in temp:
            lowercase_letter_list.append(word.lower())

    return lowercase_letter_list
```

### Linguistic Dictionary Tool

- before getting too hot and heavy, I want to make a tool that reads a given article and makes a dataframe for the words and their count

```python
text = "hello this is alex hello this is michael this is"
word_counter = {}

# establish the words present in the text
text.split(' ')
for word in text.split(' '):
    word_counter[word] = 0 

print(word_counter)

# count number of instances in each word
for word in text.split(' '):
    word_counter[word] += 1

df = pd.DataFrame(word_counter, index=[1])
df
```
